```python
# install pulp library
!pip install pulp

# import pulp library
from pulp import *

# Define decision variables representing the quantity transported on each route
xa1  = LpVariable('xa1', lowBound=0)  # Route from production A to warehouse 1
xa2  = LpVariable('xa2', lowBound=0)  # Route from production A to warehouse 2
xa3  = LpVariable('xa3', lowBound=0)  # Route from production A to warehouse 3
xb1  = LpVariable('xb1', lowBound=0)  # Route from production B to warehouse 1
xb2  = LpVariable('xb2', lowBound=0)  # Route from production B to warehouse 2
xb3  = LpVariable('xb3', lowBound=0)  # Route from production B to warehouse 3
xc1  = LpVariable('xc1', lowBound=0)  # Route from production C to warehouse 1
xc2  = LpVariable('xc2', lowBound=0)  # Route from production C to warehouse 2
xc3  = LpVariable('xc3', lowBound=0)  # Route from production C to warehouse 3
xd1  = LpVariable('xd1', lowBound=0)  # Route from production D to warehouse 1
xd2  = LpVariable('xd2', lowBound=0)  # Route from production D to warehouse 2
xd3  = LpVariable('xd3', lowBound=0)  # Route from production D to warehouse 3

x1d2 = LpVariable('x1d2', lowBound=0)  # Route from warehouse 1 to distribution 2
x1d3 = LpVariable('x1d3', lowBound=0)  # Route from warehouse 1 to distribution 3
x1d4 = LpVariable('x1d4', lowBound=0)  # Route from warehouse 1 to distribution 4
x1d5 = LpVariable('x1d5', lowBound=0)  # Route from warehouse 1 to distribution 5
x1d6 = LpVariable('x1d6', lowBound=0)  # Route from warehouse 1 to distribution 6
x2d2 = LpVariable('x2d2', lowBound=0)  # Route from warehouse 2 to distribution 2
x2d3 = LpVariable('x2d3', lowBound=0)  # Route from warehouse 2 to distribution 3
x2d4 = LpVariable('x2d4', lowBound=0)  # Route from warehouse 2 to distribution 4
x2d5 = LpVariable('x2d5', lowBound=0)  # Route from warehouse 2 to distribution 5
x2d6 = LpVariable('x2d6', lowBound=0)  # Route from warehouse 2 to distribution 6
x3d2 = LpVariable('x3d2', lowBound=0)  # Route from warehouse 3 to distribution 2
x3d3 = LpVariable('x3d3', lowBound=0)  # Route from warehouse 3 to distribution 3
x3d4 = LpVariable('x3d4', lowBound=0)  # Route from warehouse 3 to distribution 4
x3d5 = LpVariable('x3d5', lowBound=0)  # Route from warehouse 3 to distribution 5
x3d6 = LpVariable('x3d6', lowBound=0)  # Route from warehouse 3 to distribution 6

# Initialize the model as a minimization problem
TSP_MNC = LpProblem("Transshipment of MNC company", LpMinimize)

# Define the objective function to minimize the total transportation cost
# Each term represents the cost for transporting goods along a specific route
TSP_MNC += (4*xa1 + 3*xa2 + 5*xa3 + 6*xb1 + 2*xb2 + 4*xb3 + 5*xc1 + 4*xc2
            + 3*xc3 + 7*xd1 + 5*xd2 + 6*xd3 + 3*x1d1 + 2*x1d2 + 5*x1d3
            + 4*x1d4 + 6*x1d5 + 7*x1d6 + 1*x2d1+ 4*x2d2 + 2*x2d3 + 3*x2d4
            + 5*x2d5 + 6*x2d6 + 5*x3d1 + 3*x3d2 + 4*x3d3
            + 2*x3d4+ 3*x3d5 + 4*x3d6), "Total_Transportation_Cost"

 # Add supply constraints for teh productino facility
# These constraints ensure that the total quantity shipped from each source meets the supply
TSP_MNC += (xa1 + xa2 + xa3 == 70), "Supply_Constraint_A"
TSP_MNC += (xb1 + xb2 + xb3 == 90), "Supply_Constraint_B"
TSP_MNC += (xc1 + xc2 + xc3 == 80), "Supply_Constraint_C"
TSP_MNC += (xd1 + xd2 + xd3 == 60), "Supply_Constraint_D"

#if DC1 is removed

TSP_MNC += (x1d2 + x2d2 + x3d2 == 50), "Demand_Constraint_distribution_2"
TSP_MNC += (x1d3 + x2d3 + x3d3 == 70), "Demand_Constraint_distribution_3"
TSP_MNC += (x1d4 + x2d4 + x3d4 == 80), "Demand_Constraint_distribution_4"
TSP_MNC += (x1d5 + x2d5 + x3d5 == 60), "Demand_Constraint_distribution_5"
TSP_MNC += (x1d6 + x2d6 + x3d6 == 40), "Demand_Constraint_distribution_6"


# Add flow balance constraints at the transshipment nodes
# These constraints ensure that the amount entering a transshipment node equals the amount exiting
TSP_MNC += (xa1 + xb1 + xc1 + xd1  - x1d2 - x1d3 -x1d4 - x1d5 - x1d6 == 0), "Flow_Balance_warehouse_1"
TSP_MNC += (xa2 + xb2 + xc2 + xd2  - x2d2 - x2d3 -x2d4 - x2d5 - x2d6 == 0), "Flow_Balance_warehouse_2"
TSP_MNC += (xa3 + xb3 + xc3 + xd3  - x3d2 - x3d3 -x3d4 - x3d5 - x3d6 == 0), "Flow_Balance_warehouse_3"

# Solve the problem using PuLP's solver
TSP_MNC.solve()

# Print the solution: optimal transportation quantities for each route
print(TSP_MNC.name)
for v in TSP_MNC.variables():
    print(f"{v.name} = {v.varValue}")

# Print the total minimized transportation cost
print("Value of Objective Function =", value(TSP_MNC.objective))

```
